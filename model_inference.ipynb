```python
from ultralytics import YOLOv10

MODEL_PATH = "./yolov10/yolov10n.pt"
IMG_PATH = "./images/hcmc_street.jpg"

model = YOLOv10(MODEL_PATH)

result = model(source=IMG_PATH)[0]
result.save("./images/HCMC_Street_predict.png")
#print(result)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load and display the original image
original_image = Image.open("./images/hcmc_street.jpg")
predicted_image = Image.open("./images/HCMC_Street_predict.png")

# Display the images side by side
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Original image
ax[0].imshow(original_image)
ax[0].set_title("Original Image")
ax[0].axis("off")

# Predicted image
ax[1].imshow(predicted_image)
ax[1].set_title("Predicted Image")
ax[1].axis("off")

plt.show()
```
